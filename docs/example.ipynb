# Create a text file

```python
quote = "Insanity is doing the same thing over and over and expecting different results."
with open("einstein.txt", "w", encoding="utf-8") as file:
    file.write(quote)
```

# Count words

```python
counts = count_words("einstein.txt")
print(counts)
```
